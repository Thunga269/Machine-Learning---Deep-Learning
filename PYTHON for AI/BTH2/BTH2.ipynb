1. Hiển thị toàn bộ dữ liệu

```python
import pandas as pd
df = pd.read_csv('./BTH2.csv')
df
```

2. Điền các giá trị theo phương pháp trung bình

```python
df1 = df.copy()
df1.isna().sum()
```

```python
df1['Age'].fillna(round(df['Age'].mean()), inplace=True)
df1['Income'].fillna(round(df['Income'].mean()), inplace=True)
df1
```

3. Hiển thị bản ghi Income > 70000 và Age > 50

```python
df2 = df1.copy()
df2[(df2['Age']>50) & (df2['Income'] > 70000)]
```

4. Chuẩn hóa dữ liệu theo Min, Max

```python
df3 = df1.copy()
df3
```

```python
age_min = df3.Age.min()
age_max = df3.Age.max()
income_min = df3.Income.min()
income_max = df3.Income.max()
```

```python
df3.Age = (df3.Age - age_min)/(age_max-age_min)
df3.Income = (df3.Income - income_min)/(income_max-income_min)
```

```python
df3
```

5. Liên tục hóa thuộc tính Region

```python
df4 = df3.copy()
#kiểm tra thông tinh "Region"
df4.Region.unique()
```

```python
#* Đưa trường "Region" về dạng One-Hot Encoder
dummies = pd.get_dummies(df['Region'])
display(dummies)
```

```python
df4 = pd.concat([df4,dummies],axis=1)
df4.drop('Region',axis=1,inplace=True)
df4
```

6. Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính ONline Shopper dựa trên các thuộc tính khác

```python
df5 = df4.copy()
df5['Online Shopper'].replace({"No":0,"Yes":1},inplace=True)
df5
```

```python
X = df5.drop("Online Shopper", axis=1)
y = df5["Online Shopper"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lireg = LogisticRegression()
lireg.fit(X_train,y_train)
```

```python
print("Predicted label: {}, True label: {}".format(lireg.predict(X_test), y_test.values))
```

```python
lireg.score(X_test,y_test)
```
