```python
import numpy as np
import matplotlib.pyplot as plt
#dinh nghia ham loss
def Loss(theta):
  return theta**2 - 4*theta + 10
theta = np.arange(-4, 8, 0.1)
loss = Loss(theta)
plt.plot(theta, loss)
plt.show()
#dinh nghia dao ham cua ham loss
def Derloss(theta):
  return 2*theta - 4
#khoi tao tham so cua minh va cac tham so khac cua thuat toan
theta = -10
alpha = 0.01
eps = 0.0001
#Lap
#Cap nhat tham so
#tinh diem dung
while True:
  theta = theta - alpha * Derloss(theta)
  if abs(Derloss(theta) < eps):
    break
#in ket qua
print("Gia tri nho nhat cua ham loss: ", Loss(theta))
print("Gia tri nho nhat dat duoc khi: ", theta)
```
